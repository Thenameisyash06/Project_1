# Evaluating relevancy of Non-Numeric features using chi-square test

```python
from scipy.stats import chi2_contingency
non_numeric_columns = df.select_dtypes(include=['object']).columns

# Evaluate relevance using chi-square test
for column in non_numeric_columns:
    contingency_table = pd.crosstab(df[column], df['SalePrice'])
    chi2, p_value,_,_= chi2_contingency(contingency_table)
    print(f"Feature: {column}, Chi-square: {chi2}, p-value: {p_value}")
```
